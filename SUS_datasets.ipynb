```python

import pandas as pd
from google.colab import files

#Upload csv
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

#Strip whitespace from column names
df.columns = df.columns.str.strip()

print("SUS Dataset")
print(df)

# SUS Question Columns
sus_columns = [f"Q{i}" for i in range(1, 11)]
print(sus_columns)

# Calculate SUS score
def calculate_sus_score(row):
    score = 0

    # Odd or even items
    for i, col in enumerate(sus_columns, start=1):
        response = row[col]

        # Odd item
        if i % 2 == 1:
            score += response - 1

        # Even item
        else:
            score += 5 - response

    return score * 2.5


#Display computed SUS score
df["sus_score"] = df.apply(calculate_sus_score, axis=1)
print("\nIndividual SUS score")
if "USER" in df.columns:
    print(df[["USER", "sus_score"]])
else:
    print(df[["sus_score"]])

#Descriptive Statistics
mean_sus = df["sus_score"].mean()
median_sus = df["sus_score"].median()
var_sus = df["sus_score"].var()
std_sus = df["sus_score"].std()
cv_sus = (std_sus / mean_sus) * 100

print("\n Descriptive Statistics")
print(f"Mean {mean_sus:.3f}")
print(f"Median {median_sus:.3f}")
print(f"Variance {var_sus:.3f}")
print(f"Standard Deviation {std_sus:.3f}")
print(f"Coefficient of Variation {cv_sus:.3f}")

#SUS Interpretation
if mean_sus >= 72.5:
  acceptability="Acceptable"
elif mean_sus >= 51.7:
  acceptability="Marginal"
else:
  acceptability="Not Acceptable"

print(f"Acceptability: {acceptability}")

```
